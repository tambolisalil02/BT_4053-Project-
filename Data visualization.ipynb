```python
# Importing data to pd csv
data = pd.read_csv(r'E:\Data Science Projects\Fifa 21 data analysis\dataset\players_21.csv')
data.head(20)
```

```python
fill_values = data.loc[:,['pace', 'shooting', 'passing', 'dribbling', 'defending','physic']]

for i in fill_values.columns:
    data[i].fillna(data[i].mean(),inplace=True)
```

### Plotting based on Height

```python
plt.figure(figsize = (22,8))
ax = sns.countplot(x ='height_cm', data = data)
ax.set_title(label = 'Count of players based on Height', fontsize=20)
ax.set_xlabel(xlabel = 'Heights in cm', fontsize=16)
ax.set_ylabel(ylabel = 'Count',fontsize=16)
plt.show()
```

### Plot on Dribbling vs Height

```python
plt.figure(figsize = (22,8))
plt.xlabel('Height', fontsize=20)
plt.ylabel('Dribbling', fontsize=20)
plt.title ('Dribbling vs Height', fontsize=25)

sns.barplot(x='shooting' , y='height_cm', data = data.sort_values('height_cm',inplace=False),alpha=1)
```

### Weight vs Height

```python
plt.figure(figsize=(22,8))
plt.xlabel('Weight',fontsize = 20)
plt.ylabel('Height',fontsize=20)
plt.title('Weight vs Height Chart')
sns.barplot(x='height_cm', y='weight_kg', data = data.sort_values('height_cm'), alpha=1)
```

### To Show Different Work rate of players 

```python
plt.figure(figsize=(15,7))
#plt.style.use('_classic_test')

sns.countplot(x='work_rate',data=data, palette='hls')
plt.title('Different work rates of the Players Participating in the FIFA 2021', fontsize = 20)
plt.xlabel('Work rates associated with the players', fontsize = 16)
plt.ylabel('count of Players', fontsize = 16)
plt.show()
```

## Messi Vs Ronaldo

```python
data.columns
```

```python
skills = ['pace',
 'shooting',
 'passing',
 'dribbling',
 'defending',
 'physic',
 'player_traits',
 'attacking_crossing',
 'attacking_finishing',
 'attacking_heading_accuracy',
 'attacking_short_passing',
 'attacking_volleys',
 'skill_dribbling',
 'skill_curve',
 'skill_fk_accuracy',
 'skill_long_passing',
 'skill_ball_control',
 'movement_acceleration',
 'movement_sprint_speed',
 'movement_agility',
 'movement_reactions',
 'movement_balance',
 'power_shot_power',
 'power_jumping',
 'power_stamina',
 'power_strength',
 'power_long_shots']
```

```python
Messi = data.loc[data['short_name']=='L. Messi']
Messi = pd.DataFrame(Messi,columns=skills)

Ronaldo = data.loc[data['short_name']=='Cristiano Ronaldo']
Ronaldo = pd.DataFrame(Ronaldo,columns = skills)

plt.figure(figsize=(15,8))
sns.pointplot(data=Messi,color='blue')
sns.pointplot(data=Ronaldo, color='red')
plt.xticks(rotation=90)
plt.xlabel('Skills',fontsize=20)
plt.ylabel('Skill value',fontsize=20)
plt.title('Messi vs Roanldo', fontsize=25)
plt.grid()
```

### Top 5 nations with overall best players

```python
top_nations = data.groupby(['nationality'])['overall'].max().sort_values(ascending=False).head(10)
top_nations
```

### Top 5 clubs with overall Best Players

```python
top_clubs = data.groupby(['club_name'])['overall'].max().sort_values(ascending=False).head(10)
top_clubs
```

### Age Distribution of Players in the clubs

```python
top_club_names = ('FC Barcelona','Juventus','Real Madrid','Manchester City','Paris Saint-Germain')
clubs = data.loc[data['club_name'].isin(top_club_names) & data['age']]

fig,ax = plt.subplots()
fig.set_size_inches(10,10)
ax = sns.violinplot(x="club_name", y='age',data=clubs)
ax.set_title(label='Age distribution in the top 5 clubs', fontsize=25)
plt.xlabel('Clubs', fontsize=20)
plt.ylabel('Age', fontsize=20)
plt.grid()
```
